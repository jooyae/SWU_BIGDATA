```python
import csv
import matplotlib.pyplot as plt

f = open('daejeontempdata.csv')
data = csv.reader(f)
next(data)
low = []
high = []

y = input('당신의 출생년도를 입력하세요: ')
m = input('당신의 출생월을 입력하세요(입력형식:mm): ')
d = input('당신의 출생일을 입력하세요(입력형식:dd): ')

for row in data :
    if row[-1] != ''and row [-2] != '':
        date = row[0].split('-')
    if y <= date[0] :
        if date [1] == m and date[2] ==d:
            high.append(float(row[-1])) #최고 기온 저장 
            low.append(float(row[-2])) #최저 기온 저장 
plt.rc('font',family = 'Malgun Gothic') 
plt.rcParams['axes.unicode_minus'] = False
plt.title('주예 출생일부터의 대전 기온 변화 그래프')
plt.plot(high, 'red', label = 'high') 
plt.plot(low,'skyblue', label = 'low')
plt.legend() #범례 표시
plt.show() #그래프 표시 

plt.hist(low, bins = 100, color = 'b', label =  "Low")
plt.hist(high, bins = 100, color = 'r', label = "High")

plt.legend() #범례 표시
plt.show() #그래프 표시 
plt.savefig('jooyae') #png 파일 저장 
```
