```python
import numpy as np
import csv
import matplotlib.pyplot as plt
f = open('bigdatathreee.csv')
data = csv.reader(f)
next(data)
data = list(data)

name = input('인구 구조가 알고 싶은 지역의 이름을 입력해주세요 :')
mn = 1
result_name =''
result = 0
for row in data: 
    if name in row[0] : 
        home = np.array(row[101:], dtype= int) / int(row[2])
for row in data:
    away = np.array(row[101:], dtype =int)/int(row[2])
    s = np.sum((home-away)**2)
    
    if s < mn and name not in row[0]:
        mn = s
        result_name = row[0]
        result = away 

plt.style.use('ggplot')
plt.figure(figsize = (10,5), dpi = 300)
plt.rc('font', family = 'Malgun Gothic')
plt.title(name + '지역과 가장 비슷한 지역의 여성 인구 구조')
plt.plot(home, label = name)
plt.plot(away, label = result_name)
plt.legend()
plt.show()
```
