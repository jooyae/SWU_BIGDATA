```python
import csv
import pandas as pd
import os

f = open('daejeonpeople2009.csv')
d = open('daejeonpeople2018.csv')
data = csv.reader(f)
result = []
datatwo = csv.reader(d)
resulttwo = []

name = input('대전 지역의 연령별 인구 수를 알고싶은 동단위를 입력해주세요:')
for row in data:
    if name in row[0]:
        for i in row[3:] :
            result.append(int(i))
for row in datatwo:
    if name in row[0]:
        for i in row[3:] :
            resulttwo.append(int(i))

            
import matplotlib.pyplot as plt                        
plt.style.use('ggplot')
plt.rc('font', family = 'Malgun Gothic')
plt.title(name + '대전의 인구 구조')
plt.plot(result)
plt.plot(resulttwo)
plt.show() 
```
